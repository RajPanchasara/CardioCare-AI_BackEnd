## Import library

```python
import pandas as pd
```

## Read Data File

```python
df = pd.read_csv("cardio_train.csv", sep=';')
df
```

## Give String value to gender column

```python
df["gender"] = df["gender"].map({1: 'Male', 2: 'Female'})
df
```

## Convert number of day to year in age cloumn

```python
df["age"] = (df["age"]/365.25).round()
df
```

## Perform Dummy variable trap in gender cloumn

```python
df = pd.get_dummies(df, columns=['gender'], drop_first=True, dtype=bool)
df
```

## BMI Feature Engineering

```python
df['height_m'] = df['height'] / 100
df['bmi'] = df['weight'] / (df['height_m'] ** 2)
df
```

## Count missing values

```python
df = df.drop_duplicates()
df.shape
```

## Remove Duplicates

```python
print(df.isnull().sum())
```

## Outlier Detection & Removal (IQR Method)

```python
columns_to_check = ['height', 'weight', 'ap_hi', 'ap_lo']

Q1 = df[columns_to_check].quantile(0.25)
Q3 = df[columns_to_check].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

cleaned_df = df[
    ~((df[columns_to_check] < lower_bound) |
      (df[columns_to_check] > upper_bound)).any(axis=1)
]
```

```python
print(f"Total Removed rows : {df.shape[0]-cleaned_df.shape[0]}")
```

## Save new Apply

```python
cleaned_df.to_csv("cardio_train_clean.csv", index=False)
```
