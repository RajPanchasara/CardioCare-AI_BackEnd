## Import library

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## Import various algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

## 1. Load Clean Dataset

```python
df = pd.read_csv("cardio_train_clean.csv")
df.head()
```

## Split Features and Target

```python
X = df.drop("cardio", axis=1)
y = df["cardio"]
```

```python
X
```

```python
y
```

## Train–Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=18,
    stratify=y
)
```

```python
results = []
```

## K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)
acc = accuracy_score(y_test, y_pred_knn)
```

```python
name = 'K-Nearest Neighbors'

print("KNN Accuracy:", acc*100)

results.append({
        "Algorithm": name,
        "Accuracy": acc
    })
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
acc = accuracy_score(y_test, y_pred_dt)
```

```python
name = 'Decision Tree'

print("Decision Tree Accuracy:", acc*100)

results.append({
        "Algorithm": name,
        "Accuracy": acc
    })
```

## Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(X_train, y_train)

y_pred_svm = svm.predict(X_test)
acc = accuracy_score(y_test, y_pred_svm)
```

```python
name = 'Support Vector Machine'

print("SVM Accuracy:", acc*100)

results.append({
        "Algorithm": name,
        "Accuracy": acc
    })
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)

y_pred_nb = nb.predict(X_test)
acc = accuracy_score(y_test, y_pred_nb)
```

```python
name = 'Naive Bayes'

print("Naive Bayes Accuracy:", acc*100)

results.append({
        "Algorithm": name,
        "Accuracy": acc
    })
```

## Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_train, y_train)

y_pred_gb = gb.predict(X_test)
acc = accuracy_score(y_test, y_pred_gb)
```

```python
name = 'Gradient Boosting'

print("Gradient Boosting Accuracy:", acc*100)

results.append({
        "Algorithm": name,
        "Accuracy": acc
    })
```
