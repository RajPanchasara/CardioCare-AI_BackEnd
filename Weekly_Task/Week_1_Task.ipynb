<h2>Add New Cloumn "Age_Year" for age in year</h2>

<h2>Ratio of number of people to age</h2>

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Age_Year'],color='skyblue',edgecolor='black', bins=35)
plt.show()
```

<h2>Histogram of height</h2>

```python
plt.figure(figsize=(10, 5))
plt.title('Weight Distribution')
sns.histplot(df['height'],color='skyblue',edgecolor='black', bins=30)
plt.show()
```

<h2>Histogram of weight</h2>

```python
plt.figure(figsize=(10, 5))
plt.hist(df['weight'], color='skyblue', edgecolor='black', bins=30)
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.title('Weight Distribution')
plt.show()
```

<h2>Counts of gender</h2>

```python
print(df['gender'].value_counts())
df['gender'].value_counts().plot(kind='bar',color='skyblue',edgecolor='black')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

<h2>Cholesterol levels frequency</h2>

```python
print(df['cholesterol'].value_counts())
df['cholesterol'].value_counts().sort_index().plot(kind='bar',color='skyblue',edgecolor='black')
plt.xlabel('Cholesterol level')
plt.ylabel('Count')
plt.title('Cholesterol Distribution')
plt.show()
```

<h2>Glucose levels frequency</h2>

```python
print(df['gluc'].value_counts())
df['gluc'].value_counts().sort_index().plot(kind='bar',color='skyblue',edgecolor='black')
plt.xlabel('Glucose level')
plt.ylabel('Count')
plt.title('Glucose Distribution')
plt.show()
```

<h2>Smokers vs non-smokers</h2>

```python
print(df['smoke'].value_counts())
df['smoke'].value_counts().plot(kind='bar',color='skyblue',edgecolor='black')
plt.xlabel('(0 = non-smokers , 1 = Smokers)')
plt.ylabel('Count')
plt.title('Smoking Status Distribution')
plt.show()
```

<h2>Alcohol consumers vs non-consumers</h2>

```python
print(df['alco'].value_counts())
df['alco'].value_counts().plot(kind='bar',color='skyblue',edgecolor='black')
plt.xlabel('(0 = consumers, 1 = non-consumers)')
plt.ylabel('Count')
plt.title('Alcohol Consumption Distribution')
plt.colorbar()
plt.show()
```

<h2>Distribution of diastolic pressure (ap_lo)</h2>

```python
plt.hist(df['ap_lo'], color='skyblue',edgecolor='black', bins=60)
plt.xlabel('Diastolic BP (ap_lo)')
plt.ylabel('Frequency')
plt.title('ap_lo Distribution')
plt.show()
```

<h2>Distribution of systolic pressure (ap_hi)</h2>

```python
plt.hist(df['ap_hi'],color='skyblue',edgecolor='black', bins=55)
plt.xlabel('Systolic BP (ap_hi)')
plt.ylabel('Frequency')
plt.title('ap_hi Distribution')
plt.show()
```

## Alcohol vs Cardio Rate

```python
alco_cardio = pd.crosstab(df['alco'], df['cardio'], normalize='index')
print(alco_cardio)

alco_cardio.plot(kind='bar', edgecolor='black', stacked=True)
plt.title('Alcohol vs Cardio Rate')
plt.xlabel('Alco')
plt.ylabel('Proportion')
plt.show()
```
